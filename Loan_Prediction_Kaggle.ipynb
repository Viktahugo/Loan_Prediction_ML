```python
def explore_object_type(dataset,feature_name):
    if dataset[feature_name].dtype == 'object':
        print(dataset[feature_name].value_counts())
    else:
        print('Feature is not an Object data type')
```

```python
explore_object_type(df,'Gender')
```

```python
explore_object_type(df,'ApplicantIncome')
```

```python
df_columns = df.columns
```

```python
for featureName in df_columns:
    if df[featureName].dtype == 'object':
        print('\n"' + str(featureName) + '\'s" Values with count are:')
        explore_object_type(df,featureName)
```
